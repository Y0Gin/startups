<h2>–ü—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∞ –¥–∞–Ω–Ω—ã—Ö</h2>

````python
def data_info(data, data_name):
    print("–ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ —Å—Ç–æ–ª–±—Ü–∞—Ö —Ç–∞–±–ª–∏—Ü—ã ", data_name)
    display(data.info())
    print("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–µ–∑–∞–ø–æ–ª–Ω–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤ –ø–æ–ª—è—Ö:")
    print(data.isna().sum())
    dupl = data.duplicated().sum()
    print("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥—É–±–ª–∏–∫–∞—Ç–æ–≤:", dupl)  
````

````python
def dupl_processing(data):
    data = data.drop_duplicates()
    print("–í—ã–ø–æ–ª–Ω–µ–Ω–æ —É–¥–∞–ª–µ–Ω–∏–µ –¥—É–±–ª–∏–∫–∞—Ç–æ–≤")     
    print("–ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ —Å—Ç–æ–ª–±—Ü–∞—Ö —Ç–∞–±–ª–∏—Ü—ã ")
    display(data.info())
````

````python
def to_lower_case(data, list_cols):
    for col in list_cols:
        data[col] = data[col].apply(str.lower)
    return data
````

````python
def col_info(data, list_cols):
    for col in list_cols:
        print(data[col].value_counts())
    return data
````

<font color='blue'><b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞: </b></font> ‚úîÔ∏è <br>
<font color='green'>–ó–¥–æ—Ä–æ–≤–æ, —á—Ç–æ –Ω–µ –∑–∞–±—ã–≤–∞–µ—à—å –ø—Ä–æ –ø—Ä–æ–º–µ–∂—É—Ç–æ—á–Ω—ã–µ –≤—ã–≤–æ–¥—ã.</font>

<h2>–ò—Å—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑ –¥–∞–Ω–Ω—ã—Ö</h2>

````python
def numeric_exploration(data, lst_cols):
    for col in lst_cols:
        print('–°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∞—è –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –ø–æ —Å—Ç–æ–ª–±—Ü—É:', col)
        print(data[col].describe())
        plt.hist(data[col])
        plt.xlabel(col)
        plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞')
        plt.grid(True)
        plt.title('–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ —á–∞—Å—Ç–æ—Ç –ø–æ –ø—Ä–∏–∑–Ω–∞–∫—É ' + col)
        plt.show()        
        data.boxplot(column=col, vert=False)
        plt.xlabel(col)
        plt.grid(True)
        plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ –∫–≤–∞—Ä—Ç–∏–ª—è–º (—è—â–∏–∫ —Å —É—Å–∞–º–∏) –¥–ª—è –ø—Ä–∏–∑–Ω–∞–∫–∞ ' + col)
        plt.show()
````

````python
def category_exploration(data, lst_cols):
    if (len(lst_cols) == 1):
        cnt_col = "id"
    else: 
        cnt_col = lst_cols[-1]
    for col in lst_cols:
        a = data.groupby(col).agg({cnt_col:'count'}).reset_index()
        a.columns = [col, '–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ_–∑–∞–ø–∏—Å–µ–π']
        a['Percentage'] = a['–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ_–∑–∞–ø–∏—Å–µ–π'] / a['–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ_–∑–∞–ø–∏—Å–µ–π'].sum()*100
        print(a)
        a.plot(kind='bar', xlabel=col, ylabel='–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π', x=col, y='–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ_–∑–∞–ø–∏—Å–µ–π',
                              title='–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π –≤ —Ä–∞–∑—Ä–µ–∑–µ –ø—Ä–∏–∑–Ω–∞–∫–∞ ' + col, figsize=(8,8))
        cnt_col = col
````

````python
# —Å–æ–∑–¥–∞—ë–º —Å–ø–∏—Å–∫–∏ —Å –Ω–∞–∑–≤–∞–Ω–∏—è–º–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
#ohe_part_columns = ['—Ç–∏–ø_—Å–µ—Ä–≤–∏—Å–∞', '—Ä–∞–∑—Ä–µ—à–∏—Ç—å_—Å–æ–æ–±—â–∞—Ç—å']
ohe_columns = [ '—Ä–∞–∑—Ä–µ—à–∏—Ç—å_—Å–æ–æ–±—â–∞—Ç—å', '–ø–æ–ø—É–ª—è—Ä–Ω–∞—è_–∫–∞—Ç–µ–≥–æ—Ä–∏—è']
ord_columns = ['—Ç–∏–ø_—Å–µ—Ä–≤–∏—Å–∞']
num_columns = ['–º–∞—Ä–∫–µ—Ç_–∞–∫—Ç–∏–≤_6_–º–µ—Å', '–º–∞—Ä–∫–µ—Ç_–∞–∫—Ç–∏–≤_—Ç–µ–∫_–º–µ—Å', '–¥–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å', '–∞–∫—Ü–∏–æ–Ω–Ω—ã–µ_–ø–æ–∫—É–ø–∫–∏', \
               '—Å—Ä–µ–¥–Ω–∏–π_–ø—Ä–æ—Å–º–æ—Ç—Ä_–∫–∞—Ç–µ–≥–æ—Ä–∏–π_–∑–∞_–≤–∏–∑–∏—Ç','–Ω–µ–æ–ø–ª–∞—á–µ–Ω–Ω—ã–µ_–ø—Ä–æ–¥—É–∫—Ç—ã_—à—Ç—É–∫_–∫–≤–∞—Ä—Ç–∞–ª', '–æ—à–∏–±–∫–∞_—Å–µ—Ä–≤–∏—Å–∞',\
               '—Å—Ç—Ä–∞–Ω–∏—Ü_–∑–∞_–≤–∏–∑–∏—Ç', '–ø—Ä–µ–¥—ã–¥—É—â–∏–π_–º–µ—Å—è—Ü_–≤—ã—Ä—É—á–∫–∞', '–ø—Ä–µ–ø—Ä–µ–¥—ã–¥—É—â–∏–π_–º–µ—Å—è—Ü_–≤—ã—Ä—É—á–∫–∞', '—Ç–µ–∫—É—â–∏–π_–º–µ—Å—è—Ü_–≤—ã—Ä—É—á–∫–∞',
               '–ø—Ä–µ–¥—ã–¥—É—â–∏–π_–º–µ—Å—è—Ü_–º–∏–Ω—É—Ç', '—Ç–µ–∫—É—â–∏–π_–º–µ—Å—è—Ü_–º–∏–Ω—É—Ç']
````

<font color='blue'><b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞: </b></font> ‚ùå\
<font color='red'> ~~–û–±—Ä–∞—Ç–∏ –≤–Ω–∏–º–∞–Ω–∏–µ, '–ü–æ–ø—É–ª—è—Ä–Ω–∞—è_–∫–∞—Ç–µ–≥–æ—Ä–∏—è' - –Ω–µ –º–æ–∂–µ—Ç –±—ã—Ç—å –ø–æ—Ä—è–¥–∫–æ–≤—ã–º –ø—Ä–∏–∑–Ω–∞–∫–æ–º. –û–Ω –Ω–µ –≤—ã—Å—Ç—Ä–∞–∏–≤–∞–µ—Ç—Å—è –ø–æ —Ä–∞–Ω–≥—É –Ω–∏ –ø–æ –∫–∞–∫–æ–º—É —Ñ–∞–∫—Ç–æ—Ä—É.\
–ù–µ–∫–æ—Ç–æ—Ä—ã–µ –º–æ–¥–µ–ª–∏ –º–æ–≥—É—Ç —Ä–∞–±–æ—Ç–∞—Ç—å —Ç–æ–ª—å–∫–æ —Å OE (–¥–µ—Ä–µ–≤—å—è), –Ω–æ –¥–ª—è –õ–∏–Ω–µ–π–∫–∏ —ç—Ç–æ –Ω—É–∂–Ω–æ –∫–æ–¥–∏—Ä–æ–≤–∞—Ç—å –≤ OHE.~~</font>

<font color='purple'><b>–ò—Å–ø—Ä–∞–≤–∏–ª. ORD-–ø–∞–π–ø–ª–∞–π–Ω –Ω–∏–∂–µ –∑–∞–∫–æ–º–º–µ–Ω—Ç–∏—Ä–æ–≤–∞–ª.</b></font>


<font color='blue'><b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ 2: </b></font> ‚ùå\
<font color='red'> ~~–ü–æ –∑–∞–¥–∞–Ω–∏—é –Ω—É–∂–Ω–æ –¥–≤–∞ –≠–Ω–∫–æ–¥–µ—Ä–∞.\
–ú–æ–∂–Ω–æ –ø–µ—Ä–µ–¥–∞–≤–∞—Ç—å –≤ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞—Ö –¥–ª—è LR, KNN –∏ SVC - OHE, –∞ –¥–ª—è –î–µ—Ä–µ–≤–∞ OE,\
a –º–æ–∂–Ω–æ –∑–∞—Å—É–Ω—É—Ç—å –≤ ord_columns –∫–∞–∫–æ–π-–Ω–∏–±—É–¥—å –±–∏–Ω–∞—Ä–Ω—ã–π –ø—Ä–∏–∑–Ω–∞–∫, –Ω–∞–ø—Ä–∏–º–µ—Ä '—Ç–∏–ø_—Å–µ—Ä–≤–∏—Å–∞'.~~ </font>

<font color='purple'><b>–ò—Å–ø—Ä–∞–≤–∏–ª. –ó–∞—Å—É–Ω—É–ª —Ç–∏–ø —Å–µ—Ä–≤–∏—Å–∞.</b></font>

<font color='blue'><b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ 3: </b></font> ‚úîÔ∏è\
<font color='green'> üëç</font>

````python
# —Å–æ–∑–¥–∞—ë–º –ø–∞–π–ø–ª–∞–π–Ω –¥–ª—è –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏–∑ —Å–ø–∏—Å–∫–∞ ohe_columns: –∑–∞–ø–æ–ª–Ω–µ–Ω–∏–µ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –∏ OHE-–∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–µ
ohe_pipe = Pipeline(
    [('simpleImputer_ohe', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
     ('ohe', OneHotEncoder(drop='first', sparse_output=False))
    ]
    )

#ord_pipe = Pipeline(
#    [('simpleImputer_before_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
#    ('ord', OrdinalEncoder(categories=[['—Ç–æ–≤–∞—Ä—ã –¥–ª—è –¥–µ—Ç–µ–π', '–¥–æ–º–∞—à–Ω–∏–π —Ç–µ–∫—Å—Ç–∏–ª—å', '–∫—É—Ö–æ–Ω–Ω–∞—è –ø–æ—Å—É–¥–∞', '–∫–æ—Å–º–µ—Ç–∏–∫–∞ –∏ –∞–∫—Å–µ—Å—É–∞—Ä—ã',\
#                     '—Ç–µ—Ö–Ω–∏–∫–∞ –¥–ª—è –∫—Ä–∞—Å–æ—Ç—ã –∏ –∑–¥–æ—Ä–æ–≤—å—è', '–º–µ–ª–∫–∞—è –±—ã—Ç–æ–≤–∞—è —Ç–µ—Ö–Ω–∏–∫–∞ –∏ —ç–ª–µ–∫—Ç—Ä–æ–Ω–∏–∫–∞']], 
#                handle_unknown='use_encoded_value', unknown_value=np.nan)),
#      ('simpleImputer_after_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent'))
#     ]
#)

ord_pipe = Pipeline(
    [('simpleImputer_before_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
    ('ord', OrdinalEncoder(categories=[['–ø—Ä–µ–º–∏—É–º', '—Å—Ç–∞–Ω–¥–∞—Ä—Ç']], 
                handle_unknown='use_encoded_value', unknown_value=np.nan)),
      ('simpleImputer_after_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent'))
     ]
)

num_pipe = Pipeline(
    [('simpleImputer_numb', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
     ('sc', MinMaxScaler()),
     ('simpleImputer_numa', SimpleImputer(missing_values=np.nan, strategy='most_frequent'))
    ]
    )

# —Å–æ–∑–¥–∞—ë–º –æ–±—â–∏–π –ø–∞–π–ø–ª–∞–π–Ω –¥–ª—è –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∏ –¥–∞–Ω–Ω—ã—Ö
data_preprocessor = ColumnTransformer(
    [('ohe', ohe_pipe, ohe_columns),
     ('ord', ord_pipe, ord_columns),
     ('num', num_pipe, num_columns)
    ], 
    remainder='passthrough'
)

# —Å–æ–∑–¥–∞–µ–º –∏—Ç–æ–≥–æ–≤—ã–π –ø–∞–π–ø–ª–∞–π–Ω: –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö –∏ –º–æ–¥–µ–ª—å
pipe_final = Pipeline(
    [
        ('preprocessor', data_preprocessor),
        ('models', DecisionTreeClassifier(random_state=RANDOM_STATE))
    ]
)
````

````python
param_grid = [
    # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ DecisionTreeClassifier()
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(2,5),
        'models__max_features': range(2,5),
        'preprocessor__num': [ MinMaxScaler(), RobustScaler(), StandardScaler(), 'passthrough']

    },
    # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ KNeighborsClassifier() 
    {
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': range(2,5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), 'passthrough']
    },

    # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ LogisticRegression()
    {
        'models': [LogisticRegression(
            random_state=RANDOM_STATE, 
            solver='saga', 
            penalty='l1'
        )],
        'models__C': np.arange(0.1,3.00,0.1),
        'models__solver': ['saga', 'liblinear'],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), 'passthrough']
        
    },
    
    # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ SVC()
    {
        'models': [SVC(
            random_state=RANDOM_STATE,
            probability= True,
            kernel='sigmoid'
        )],
        'models__kernel': ['sigmoid', 'rbf', 'linear'],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), 'passthrough']
    }
]

grid = GridSearchCV(
    pipe_final, 
    param_grid=param_grid, 
    cv=5, 
    scoring='roc_auc', 
    #random_state=RANDOM_STATE, 
    n_jobs = -1,
    error_score='raise'
)
grid.fit(X_train, y_train) 
````
